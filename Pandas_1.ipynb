## DataFrame 객체

## 데이터 병합: concat, append, merge

* 병합: 여러 개의 데이터프래임(dataframe)이나 시리즈(Series)를 결합

* concat은 행(row)과 열(column)을 모두 병합
* append는 행(row)
* merge는 열(column)

image.png

* df1, df2, df3 데이터프레임 생성

```
import pandas as pd
import numpy as np

data1 = {"id": [1,2,3,4,5],
         "gender": ['M','F', 'F', 'M', 'F'],
         "age": [21, 25,22,20,24],
         "pass": ["Yes","No", "Yes", "Yes", "No"],
         "attend": [14,16,12,16,12]}

df1 = pd.DataFrame(data1)
df1
```

```
data2 = {"id": [1,2,3,4],
         "age": [21, 25,22,21],
         "happy": [4,6,5,7],
         "TV": [1,1.5,3,2.5],
         "game": [2, 0.5, 2.5, 2]
         }

df2 = pd.DataFrame(data2)
df2
```

```
data3 = {"id": [6,7,8],
         "gender": ['M','F', 'F'],
         "age": [22, 20,24],
         "pass": ["Yes","No", "Yes"]
         }

df3 = pd.DataFrame(data3)
df3
```

#### concat(): 행과 열병합

image.png

*  행병합(axis=0)

```
df1_3 = pd.concat([df1,df3], ignore_index = True)
df1_3
```

* 열병합(axis=1)

```
df1_2 = pd.concat([df1,df2], axis=1)
df1_2
```

* 중복컬럼 제거

```
df1_2 = df1_2.loc[:,~df1_2.columns.duplicated()].copy()
df1_2
```

#### append(): 행 병합

image.png

```
df1_3 = df1.append(df3, ignore_index=True)
df1_3
```

#### merge(): 컬럼 병합

* merge: 양쪽 데이터프래임에 공통으로 가지고 있는 컬럼을 키(key)값으로 사용해 병합

* 데이터프래임을 기준(how='left', or how='right') 설정
* 어느 한쪽에라도 데이타가 없는 경우 NaN 값을 지정할지(how='outer'), 아니면 데이터를 삭제할지(how='inner') 결정
* 어떤 key 값을 이용해(on='key')을 병합할지를 결정
* df1과 df2에 중복으로 있는 변수(age) 제거

image.png

```
df1_2_outmerge = pd.merge(df1,df2, how='outer', on='id')
df1_2_outmerge
```

```
df1_2_innermerge = pd.merge(df1, df2, how='inner', on='id')
df1_2_innermerge
```

```
df1_2_innermerge = pd.merge(df1,df2, on='id').drop("age_y", axis=1)
df1_2_innermerge
```
