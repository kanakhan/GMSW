## 자연어 처리(NLP)

* 자연어(natural language)란 우리가 일상생활에서 사용하는 인간의 언어를 의미
* 컴퓨터가 처리하는 언어는 기계어(machine language).
* 자연어 처리(Natural Language Processing, NLP)는 컴퓨터가 인식할 수 있는 언어로 바꿔서 자연어를 분석해 대용량의 텍스트로부터 의미 있는 정보를 찾아내는 방법

#### 토큰화

* 텍스트를 처리하기 위해서는 텍스트(corpus)에 대해 문장이나 단어 등으로 언어의 경계를 컴퓨터에 알려주어야 하는데 그 경계는 '의미 부여'가 가능한 단위여야 함.
*'의미 단위'로 구분하는 것을 토큰화(tokenization)라고 하고 단어로 분리하면 단어 토큰화(word tokenization), 문장으로 토큰화하면 문장 토큰화(sentence tokenization)라고 함.

####  정제(cleaning)

* 특수문자나 숫자, 자음, 모음 등 분석에 불필요하거나 오히려 잡음으로 작용할 수 있는 단어를 제거


* 토큰화를 실시한 후 유의미한 단어 토큰만을 선별하고 의미 없는 단어들을 제거하기 위해 불용어(stopwords) 사전 이용


#### 임베딩(embedding, encoding)

* 분리된 단어를 컴퓨터가 처리 가능하도록 수치를 부여하는 작업을 단어 임베딩(word embedding) 또는 자연어를 기계어로 바꾸는 인코딩(encoding) 작업이라고 함.


* 임베딩 방법은 BOW(bag of words), countVectorizer, TF-IDF와 같은 고전적인 단어 임베딩에서 부터 Word2Vec, Glove, fastText, Bert 등 딥러닝을 이용한 임베딩 방식까지 다양함.
