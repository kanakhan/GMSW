# 클래스(class)

* 클래스는 틀이나 설계도
* 이 설계도나 틀에 의해서 만들어진 것들을 객체(object) 혹은 인스턴스(instance)라고 함.
* 객체는 속성(attributes)과 기능(methods)

#### 만약 클래스를 사용하지 않는다면 얼마나 번거로운지 실제 게임 캐릭터(마법사)를 만들어 보면 쉽게 이해가 될 것


### 책의 틀(저자, 제목, 출판사, 발행인)이라는 클래스를 만들어 놓고 그 틀에 맞춰 책들을 계속 찍어 낸다(인스턴스)


image.png

### 클래스의 기본

*   책(인스턴트)이 저자, 제목, 출판사, 발생 일을 속성으로 갖도록 클래스에 틀을 짜 놓고 계속해서 찍어 냄

### 클래스의 메서드 정의

* 클래스 안에 함수(def)를 정의 내려서 이를 메서드로 사용하는 것이 가장 일반적
* 메서드를 사용하기 위해서는 클래스 자체를 받을 self를 쓰게 되는데 이는 위의 예제에서 클래스 자체를 받는 book(=Book()) 인스턴스를 의미



image.png

### 클래스 속성과 인스턴스 속성

* 만약 출판 부수를 넣고 싶다면, 이는 각각의 책(instance)은 1로 의미가 없음.
*  따라서 출판 부수는 인스턴트의 속성이 아니라 클래스가 가져야 할 속성 그럼 출판 부수(count) * 클래스 속성은 Book, 인스턴트 속성은 book으로 구분

## 클래스 매직 메서드(Class Magic Methods)

image.png

* 먼저 __init__() 메서드는 클래스의 속성들을 초기화하라는 의미로 그렇기 때문에 객체나 인스턴스를 생성할 때 맨 처음으로 초기화하는 작업을 __init__에서 진행
* __str__() 메서드는 인스턴스를 출력하라는 의미

## 클래스 상속



*   클래스를 상속한다는 것은 상위 클래스에서 정의된 속성과 메서드를 하위(다음) 클래스가 그대로 물려받는다는 것을 의미

*   이렇게 되면 새로운 코드를 작성할 필요가 없고 클래스 간 계층 관계를 구분할 수 있어서 코드 관리가 매우 수월

## 메서드 오버라이딩(Method Overriding)

* 메서드 오버라이딩(Method Overriding)은 부모 클래스로부터 상속받은 특정 메서드를 자식 클래스에서 재정의하는 작업을 의미


####(예제)  자식 클래스에서는 부모 클래스에 '책에 대한 평가'를 넣음.


## super ( )

* supers는 상속의 대상인 부모 클래스를 호출하는 함수
*자식 클래스 FictionBook이 부모 클래스 Book으로부터 상속받아 부모 클래스의 개체 속성을 그대로 물려받기 위해 Book.__init__(self, author, title, publisher, date)을 사용하였다. 여기서 super( )가 이와 같이 상위 클래스를 호출하는 역할
*표현식은 super( ).__inti__(author, title, publisher, date)로 self가 빠지게 된다.
