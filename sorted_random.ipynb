### 랜덤 모듈



*   전체 집합에서 일부인 표본(sample)을 뽑아서 전체를 대표하도록 하기 위해서는 샘플을 랜덤(random) 하게 무작위로 뽑는다는 것이 가정




*  랜덤하게 뽑기 위해서는 난수(랜덤 수)가 필요
* 파이썬(Python)에서 난수를 만들기 위해서는 random 모듈을 사용. random 모듈에는 내미덤과 관련된 함수들을 모두 모아 두고 있음.




image.png

```python
from random import *
#import * -> import 안의 모든 함수를 가져옴

print(random())
print(randint(1,100))
print(uniform(1.0,10.5))
print(randrange(2,100,2))
print(randrange(10))
```

#### (예제) 1~45 중 번호 랜덤하게 뽑기

```python
print(randrange(1,46))
print(randint(1,45))

a = []
i=0

for i in range(7):
  k = randint(1,45)
  if (k in a):
    continue
  else:
    a.append(k)

print(f"로또번호는 {a} 입니다.")

```

#### (예제) random 모듈 사용: up, down 숫자 맞추기 게임

```python
myNum = randint(1,100)

while True:
  yourNum = int(input("당신의 숫자를 입력하세요: "))
  if (myNum == yourNum):
    print(f"맞았습니다!! 숫자는 {myNum}이었습니다.")
    break
  elif (myNum>yourNum):
    print("UP")
  else:
    print("DOWN")
```

## shuffle ()을 이용해 데이터 섞기

```python
li = [1,2,3,4,5]
print(choice(li))
print(sample(li,5))
print(choices(li, k=5))

fruit = ["apple", "pear", "grape", "mango", "banana"]
shuffle(fruit)
print(sorted(fruit))

```

### (예제) suffle을 이용해 로토 번호 5개 뽑기

```python
lott = list(range(1,46))
shuffle(lott)
print(lott)
print(lott[:6])
print(sorted(lott))

lotto = list(range(1,46))
shuffle(lotto)
print(lotto[:6])
```
