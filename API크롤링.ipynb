```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
import pandas as pd
import json 
import re

client_id = "lnvupc2h1nPqE8JGNIbL"
client_secret = "Gut57HSE_D"
encText = urllib.parse.quote("손흥민")

blog_df = pd.DataFrame(columns = ("title", "link", "description", "bloggername", "bloggerlink", "postdate"))

start = 1
end = 1000
display = 100

for start_idx in range(start,end,display):
  url = "https://openapi.naver.com/v1/search/blog?query=" + encText \
  + "&display=" + str(display) + "&start" + str(start_idx) # JSON 결과



  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()
  if(rescode==200):
      response_body = response.read()
      response_dict = json.loads(response_body.decode('utf-8'))
      items = response_dict['items']

      idx = 0

      for item_idx in range(0, len(items)):
        remove_tag = re.compile('<.*?>')
        title = re.sub(remove_tag, '', items[item_idx]['title']) ##컴파일을 시켜서 빈칸으로 만들고, 타이틀에 집어넣음
        link = re.sub(remove_tag, '', items[item_idx]['link'])
        description = re.sub(remove_tag, '', items[item_idx]['description'])
        bloggername = re.sub(remove_tag, '', items[item_idx]['bloggername'])
        bloggerlink = re.sub(remove_tag, '', items[item_idx]['bloggerlink'])
        postdate = re.sub(remove_tag, '', items[item_idx]['postdate'])
        blog_df.loc[idx] = [title, link, description, bloggername, bloggerlink, postdate]
        idx += 1
  else:   
      print("Error Code:" + rescode)
```
