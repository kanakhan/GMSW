## 토픽 모델링

* 토픽 모델링의 대표적인 방법으로 잠재 의미 분석(Latent Semantic Analysis)과 잠재 디리클레 할당(Latent Dirichlet Allocation) 기법이 있음.
* 여기서는 잠재 디리클레 할당(LDA)을 살펴보기로 함.

#### 잠재 디리클레 할당(LDA) 기본 가정


* 잠재 디리클레 할당(LDA)은 글을 쓰는 원리를 역추적함.
* 일반적으로 하나의 문서는 여러 토픽을 담고 있으며, 해당 토픽은 토픽을 설명할 수 있는 단어로 구성되어 있음.

* 예를 들어) 요리책은 ''돼지고기, 파, 양파, 등'으로 구성된 재료 토픽, '온도, 시간, 순서 등'으로 구성된 방법 토픽으로 구성되어 있으며, 첫 번째 페이지(문서)는 재료 70%, 방법 30%로 구성되어 있다는 것을 추론하는 것
* 이런 과정을 끝냈다면, LDA는 재료 토픽을 구성하는 단어 중 랜덤하게 70%를 뽑고, 방법을 구성하는 단어 30%를 뽑으면 요리책 첫 번째 페이지가 작성된다고 가정

#### 잠재 디리클레 할당(LDA) 목표

* 각각의 문서에 포함된 토픽의 분포(예, 문서 1은 스포츠 토픽 30%, 경제 토픽 50%, 시사 토픽 20%)

* 토픽 별로 포함된 단어 분포( 예, 스포츠 토픽: 축구 30%, 야구 50%, 배구 15%, 연봉 5%)
