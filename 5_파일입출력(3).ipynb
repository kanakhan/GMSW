### 파일 내용 읽기 : readline(), readlines(), read()

```python
f = open("file.txt", 'w')
f.write("안녕하세요 저는 트위치에서 방송을 하고있는 스트리머 케인입니다")
f.write("\n먼저 저의 말과 행동으로 인해 큰 피해를 끼치고 실망을 드린 샌드백님")
f.write("\n시청자분들께")
f.write("\n죄송합니다.")
f.close()
```

```python
!cat file.txt
```

```python
text = ("""케인인님
한
판 
마
안""")

f=open("file.txt", "w")
f.writelines(text)
f.close()

!cat file.txt
```

```python
text = input("케인의 코 크기는??: ")
f = open("file.txt", 'w')
f.write(text)
f.close()

!cat file.txt
```

```python
text = [str(i + '오옹나이스\n') for i in input("여러 줄 입력: ").split(",")]
f = open('file.txt', 'w')
f.writelines(text)
f.close()

print(text[0])
!cat file.txt
```

```python
f = open("file.txt", 'w')
line = ""
while True:
  line = input()
  if not line:
    break
  f.writelines(line + '\n')

f.close()

!cat file.txt
```

### 파일 내용 읽기 : readline(), readlines(), read()

```python
f = open('file.txt', 'r')
while True:
  line = f.readline()
  if not line:
    break
  print(line, end="")

f.close()
```

```python
f = open('file.txt', 'r')
data = f.readlines()
print(data)
f.close()
```
