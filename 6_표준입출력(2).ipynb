

*   출력문 정렬방식: ljust, rjust



```python
score = {"수학":0, "영어": 50, "무빙": 100, "큰코" : -3000}
for subject, score in score.items():
  print(subject.ljust(4), str(score).rjust(8), sep = ":")
```



*   출력문 자리수 지정 및 빈자리 채우기: zfill

```python
for num in range(1,5):
  print("대기번호 : %s" %(str(num).zfill(3)))
```



*   출력문 숫자 형식 지정



{0: +} => 숫자에 부호 삽입

{0: , } => 3 자리 마다 컴마

{0: > 10} => 10자리 확보 왼쪽 정열

{0: < 10} => 10자리 확보 오른쪽 정열

{0:_<10} => 10칸 확보하고 빈자리 _ 채움

{0: f} => 소수점 출력, {0, 2f} => 소수점 2째 자리까리(3째 자리에서 반올림)

{0:^<+10,} => 빈칸에 ^를 채움, 왼쪽 정렬, 부호 나타냄, 10자리 확보, 3자리에 컴마찍음.
[출처] 파이썬 (python) 코딩 표준 입출력 input과 print를 활용해 빅데이터 분석 마스터 하기|작성자 행복전염

```python
print("{0:<100,}".format(100000000), end = "")
print("집에가고싶어요")
```

```python
print("{0:,.2f}".format(100000000.42152))
```

```python
print("{0:^<+30}{1:+10} {2:>10}" .format(1,2,3))
```

## print() 함수내 변수 및 값 반환 






*  print() 함수는 print() 함수 안에서 특정 내용을 따로 지정하고 그 값만 변화시키면서 출력 할 수 있다. 






1.   방법1.  문자열 안에 정수, 문자열, 문자를 삽입할 경우 각각 %d, %s, %c를 사용




```python
print("안%d녕" %1000000)
```

```python
print("나는 %s 를! 했다" %'장풍')
```

```python
print(" %c먹어 %c" %('배','배'))
```



2.  방법2."{0} {1}".format(A, B) 를 활용


*   여기서 {0} 첫 번째 내용인 A를 {1}은 두 번째 내용은 B를 가리킨다.






```python
print("니 하는 플레이 보이 예전에 {} 같은데" . format("하더놈"))
print("니 하는 플레이 보이 예전에 {1} 같은데 이제 {0} 하는갑지" . format("하더놈", "얼굴까고"))
```


3. 방법 3: "{A}".format(A=B), 문자열 B를을 A변수로 지정하여 삽입


```python
print("{사람}님의 코는 {크기} 이며, {측정가능여부}이다.".format(사람='케인인', 크기 = -3000, 측정가능여부="측정불가능"))
```



4. 방법 4: 변수를 활용(f"{A}{B}"), 변수 A와 B는 외부에서 정의된 변수



```python
name = '케인인님'
age = -3000
her = '타지리'

print(f"{name}은 {age}살이고 {her}와 맺어졌다")
```

### (예제) 구구단: 각 단마다 행으로 보기
